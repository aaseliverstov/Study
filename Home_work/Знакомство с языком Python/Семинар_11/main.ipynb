Даны две функции:
f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241

Требуется:
1. Найти координаты точек пересечения
2. Построить графики функций в одной системе координат
3. Построить график функции пересечения

```python
from sympy import *

x = Symbol("x", real=True)

fx = collect((x ** 3 - 50*x), x)
f = solve(Eq(fx, 0))

gx = collect((-x ** 4 + 88*x**2 - 241), x)
g = solve(Eq(gx, 0))

fgx = collect(fx - gx, x)
fg = solve(Eq(fgx, 0))
```

Координаты точек пересечения

```python
dict = {}
count = 1
for i in fg:
    dict[count] = {'x': round(N(i), 2), 'y': round(N(-i**4 + 88*i**2 - 241), 2)}
    count += 1

for i in dict:
    print(f"x = {dict[i]['x']}, y = {dict[i]['y']}")
```

Графики функций в одной системе координат

```python
plot(fx, gx)
```

График функции пересечения

```python
plot(fgx, line_color='g')
```

Общий график пересечений функций

```python
plot(fx, gx, fgx)
```
